# Initialise

```python
#https://matplotlib.org/stable/plot_types/index.html
import matplotlib.pyplot as plt
import numpy as np
```

# Line Graph: Plot(x, y)

```python
# Choose graph style
# https://www.dunderdata.com/blog/view-all-available-matplotlib-styles
plt.style.use('default')

# make data
x = np.linspace(0, 10, 100) # Generate 100 samples within (1, 10)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots() # fig is the big Figure, ax is the subfigure in it

ax.plot(x, y, linewidth=2.0)

# Set style
# xlim sets the range, xticks sets the resolution
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Scatter Graph: Scatter(x, y)

```python
plt.style.use('default')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

# c is a color list
# Norm or vin/vmax is used to scale 'c' if 'c' is float
# The narrower the vin/vmax is, the darker the colour is
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

# np.arange returns an evenly spaced array within the range
ax.set(xlim=(0, 9), xticks=np.arange(1, 9), ylim=(0, 9), yticks=np.arange(1, 9))

plt.show()
```

# Fill Graph: Fill_between(x, y)

```python
plt.style.use('default')

# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))

# plot
fig, ax = plt.subplots()

# Fill between y1 and y2, x is only used to set the xticks
ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Square_grid_tense Graph: Imshow(x)

```python
# X and Y must be squares

plt.style.use('default')

# make data
# X and Y are used to generate the grids
X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
# Z is the value for each grid
# According to the fomula, the shape of Z is regulated by X and Y automatically
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# plot
fig, ax = plt.subplots()

ax.imshow(Z)

plt.show()
```

# Skewed_grid_tense Graph: Pcolormesh(x, y, z)

```python
# x and y can be skewed
plt.style.use('default')

# make data with uneven sampling in x
x = [-3, -2, -1.6, -1.2, -.8, -.5, -.2, .1, .3, .5, .8, 1.1, 1.5, 1.9, 2.3, 3]

# np.meshgrid is used to generate grids
# x is xticks, np.linspace is yticks
# 写的时候x与y一定要调换位置
X, Y = np.meshgrid(np.linspace(-3, 3, 128), x)
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# plot
fig, ax = plt.subplots()

# vmin, vmax指矩阵里最大最小值
ax.pcolormesh(X, Y, Z, vmin=-0.5, vmax=1.0)

plt.show()
```

# Contour: Contour(x, y, z, levels)

```python
plt.style.use('default')

# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots()

ax.contour(X, Y, Z, levels=levels)

plt.show()
```

# Fill_contour: Contourf(x, y, z, levels)

```python
plt.style.use('default')

# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# The number of levels can be set extremely large to form a high-resolution graph
levels1 = np.linspace(Z.min(), Z.max(), 7)
levels2 = np.linspace(Z.min(), Z.max(), 100)
# plot
fig, ax1 = plt.subplots()
fig, ax2 = plt.subplots()
ax1.contourf(X, Y, Z, levels=levels1)
ax2.contourf(X, Y, Z, levels=levels2)

plt.show()
```

# Box Graph: Boxplot(Data, position, styles)

```python
plt.style.use('default')

# make data:
np.random.seed(10)

# D contains data for all the columns
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# plot
fig, ax = plt.subplots()

# median: 中位数那条线
# box: 25% - 75% 中间的box
# whisker: 竖直的线
# cap: 水平表示最大最小值的线
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Error_bar Graph: Errorbar(x, y, error)

```python
plt.style.use('default')

# make data:
np.random.seed(1)
x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 1.2, 0.5]

# plot:
fig, ax = plt.subplots()

# fmt: median y value
ax.errorbar(x, y, yerr, fmt='o', linewidth=2, capsize=6)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Violin Graph: Violinplot(Data, position, styles)

```python
plt.style.use('default')

# make data:
np.random.seed(10)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))

# plot:
fig, ax = plt.subplots()

vp = ax.violinplot(D, [2, 4, 6], widths=2,
                   showmeans=False, showmedians=True, showextrema=False)
# 设置透明度
for body in vp['bodies']:
    body.set_alpha(0.9)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Spike Graph: Evenplot(Data, orientation)

```python
plt.style.use('default')

np.random.seed(1)
# x is the position
x = [2, 4, 6]
D = np.random.gamma(4, size=(3, 50))

# plot:
fig, ax = plt.subplots()

ax.eventplot(D, orientation="horizontal", lineoffsets=x, linewidth=0.75)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# 3D Hillshade Graph: Plot_surface(x, y, z)

```python
from matplotlib import cbook
from matplotlib import cm
from matplotlib.colors import LightSource
import matplotlib.pyplot as plt
import numpy as np

# Load and format data
dem = cbook.get_sample_data('jacksboro_fault_dem.npz', np_load=True)
z = dem['elevation']
nrows, ncols = z.shape
x = np.linspace(dem['xmin'], dem['xmax'], ncols)
y = np.linspace(dem['ymin'], dem['ymax'], nrows)
# Need to generate grids at first
x, y = np.meshgrid(x, y)

region = np.s_[5:50, 5:50]
x, y, z = x[region], y[region], z[region]

# Set up plot
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

ls = LightSource(270, 45)
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
rgb = ls.shade(z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, facecolors=rgb,
                       linewidth=0, antialiased=False, shade=False)
plt.show()
```
