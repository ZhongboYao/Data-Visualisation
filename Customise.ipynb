# Basic Graph

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots()

ax.contour(X, Y, Z, levels=levels)

plt.show()
```

# Annotations

```python
plt.style.use('default')

# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots()

ax.contour(X, Y, Z, levels=levels)

ax.annotate('Point A', xy=(0, 0.2),  xycoords='data',
            xytext=(0.8, 0.95), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='top',
            )

plt.show()
```

# Subplots and Titles

```python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# Set subplots and sizes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3))

ax1.contour(X, Y, Z, levels=levels)
ax2.contour(X, Y, Z, levels=levels)

# Set titles
fig.subplots_adjust(top=0.75) # 设置主副标题距离
fig.suptitle('Contour Figure', fontsize=14, fontweight='bold')
ax1.set_title('fig1')
ax2.set_title('fig2')

plt.show()
```

# Insert Text

```python
plt.style.use('default')

# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots()

ax.contour(X, Y, Z, levels=levels)

# 这两个数为坐标系位置
ax.text(-1, 2.5, r'an equation: $E=mc^2$', fontsize=12)
        
plt.show()
```

# Set Legend

```python
plt.style.use('default')

# make data
x = np.linspace(0, 10, 100) # Generate 100 samples within (1, 10)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots() # fig is the big Figure, ax is the subfigure in it

ax.plot(x, y, linewidth=2.0, label='line a')
ax.legend()
# Set style
# xlim sets the range, xticks sets the resolution
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Colour Bar

```python
from matplotlib import cm

fig, ax = plt.subplots()

contour = ax.contourf(X, Y, Z, levels=100)

# Add colorbar, make sure to specify tick locations to match desired ticklabels
# tick的值对应vmax vmin
cbar = fig.colorbar(contour, ticks=np.linspace(np.amin(Z), np.amax(Z), 10))
cbar.ax.set_yticklabels(np.linspace(np.amin(Z), np.amax(Z), 10))
```
